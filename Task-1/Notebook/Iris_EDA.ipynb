## Data Visualization

Data visualization is used to better understand the distribution, patterns, relationships, and potential outliers present in the dataset.  
Different plots are used to analyze the data from multiple perspectives, such as value distributions, variability, and feature relationships.

- Histograms help in understanding how values are spread across different ranges.  
- Boxplots are useful for identifying the spread of data and detecting outliers.  
- Scatter plots show relationships between features and highlight patterns among different species.

These visualizations provide important insights that support further analysis and model building.



## Boxplot of SepalLength & SepalWidth

```python
plt.figure(figsize=(6,5))
sns.boxplot(data=df[['SepalLengthCm','PetalLengthCm','SepalWidthCm','PetalWidthCm']].dropna())
plt.title("Boxplot of Petal Length")

plt.show()
```

## Box Plot Insights

- `PatelLengthCm` shows the highest variability, indicating strong differences among flowers species.
- `PetalWidthCm` has relatively clear value separation , making it an importanat feature for classification.
- `SepalLengthCm` displays moderate spread with overlapping values , so it is less effective for species separation.
- `SepalWidthCm` contains visible outliers, suggesting irregular observations and lower consistency.

### Conclusion
 Petal based features provide more descriminative power than sepal-based features , which explains their importance in Iris Classification Model.







## Scatter Plot   (SepalLength vs SepalWidth )

```python
sns.scatterplot(
    data=df,
    x="SepalLengthCm",
    y="SepalWidthCm",
    hue="Species"
)
plt.title("SepalLength vs SepalWidth")
plt.show()
```

##  Scatter Plot Insights

- The scatter plot shows a weak relationship between sepal length and sepal width.
- Setosa points are clearly clustered and have shorter sepal lengths with wider sepal widths.
- Versicolor and Virginica points overlap with each other, making them harder to distinguish using sepal features alone.

This indicates that sepal length and sepal width are not strong features for separating all iris species.


## Scatter Plot (Petal Length vs Petal Width)

```python
sns.scatterplot(
    data=df,
    x="PetalLengthCm",
    y="PetalWidthCm",
    hue="Species"
)
plt.title("PetalLength vs PetalWidth")
plt.show()
```

## Petal Length vs Petal Width â€“ Scatter Plot Insights

- The scatter plot shows a strong positive relationship between petal length and petal width.
- Each species forms a distinct cluster with minimal overlap.
- Setosa is clearly separated with very small petal length and width values.
- Versicolor and Virginica are also distinguishable based on petal features.

This shows that petal length and petal width are highly important for species classification.


## Histogram

```python
df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].hist(figsize=(10,8))
plt.show()
```

## Histogram Insights of Iris Features

The histogram of **Sepal Length** shows that most values lie between 5 and 6.5 cm, indicating a moderate spread of sepal length across the dataset.

The **Sepal Width** histogram shows values mostly concentrated between 2.5 and 3.5 cm, suggesting less variation compared to sepal length.

The **Petal Length** histogram shows distinct groups of values, indicating clear differences among species. This suggests petal length varies significantly and can help in classification.

The **Petal Width** histogram also shows separated ranges, with some values very small and others much larger. This confirms that petal width is an important feature for distinguishing iris species.

Overall, petal-related features show clearer distribution patterns compared to sepal features.
